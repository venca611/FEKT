Ohmův zákon

```python
import SchemDraw
import SchemDraw.elements as elm
#import matplotlib.pyplot as plt
#plt.xkcd()
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

d1 = SchemDraw.Drawing()
R = d1.add(elm.RBox(botlabel = '$R$'))
d1.labelI(R, '$U$')
d1.labelI_inline(R, botlabel = '$I$')

d1.draw()
```

$U = R\cdot I$

1. Kirchhoffův zákon (proudový):

   Proudy tekoucí z uzlu bereme s kladným znaménkem, proudy tekoucí do uzlu se záporným znaménkem.

$\qquad\sum\limits_{k} I_k = 0$

```python
d2 = SchemDraw.Drawing()
W1 = d2.add(elm.LineDot())
d2.labelI(W1, label = '$I_1$', reverse = True)
d2.push()

W2 = d2.add(elm.Line(theta = 60))
d2.labelI(W2, label = '$I_2$', reverse = True)
d2.pop()
W3 = d2.add(elm.Line(theta = -60))
d2.labelI(W3, label = '$I_3$')

d2.draw()
```

$\qquad I_1 - I_2 + I_3 = 0$

    2. Kirchhoffův zákon (napěťový):
    
    Napětí (úbytky na rezistorech, napětí zdrojů), jejichž čítací šipka má směr souhlasící se směrem oběhu kolem smyčky, bereme s kladným znaménkem, ostatní napětí se záporným znaménkem.

```python
d3 = SchemDraw.Drawing(unit = 3)
d3.add(elm.LineDot(l=1))
V1 = d3.add(elm.SourceV(d = 'right', reverse = True))
d3.labelI(V1, label = '$U_{i_1}$', arrowofst = 1, arrowlen = 1.5)
R3 = d3.add(elm.RBox(botlabel = '$R_3$'))
d3.labelI(R3, label = '$U_{R_3}$', arrowofst = 1, arrowlen = 1.5)
d3.push()

d3.add(elm.LineDot, d = 'up',reverse = True, l = 1)
d3.pop()
V2 = d3.add(elm.SourceV(d = 'right'))
d3.labelI(V2, label = '$U_{i_2}$', arrowofst = 1, arrowlen = 1.5, reverse = True)
R2 = d3.add(elm.RBox(botlabel = '$R_2$'))
d3.labelI(R2, label = '$U_{R_2}$', arrowofst = 1, arrowlen = 1.5, reverse = True)
d3.push()

d3.add(elm.LineDot(reverse = True, l = 1))
d3.pop()
W1 = d3.add(elm.LineDot('down', l=6))
d3.push()

d3.add(elm.Line(l=1))
d3.pop()
R1 = d3.add(elm.RBox('left', label = '$R_1$', tox = V1.start))
d3.labelI(R1, label = '$U_{R_1}$', arrowofst = 1.5)
d3.push()

d3.add(elm.LineDot('down', reverse = True, l = 1))
W2 = d3.add(elm.Line('up', toy = V1.start))
d3.loopI([R3, W1, R1, W2], label='$I_1$', pad=1.25)

d3.draw()
```

$\qquad U_{i_1}+U_{R_3}-U_{i_2}+U_{R_2}+U_{R_1} = 0$
